# 4 EDA of Categorical Variables

# A. Missing value treatment

```python
print('total number of categorical null values:', df[df_categorical].isna().sum().sum(), '\n')


missing_number_categorical = df[df_categorical].isna().sum().sort_values(ascending=False)
missing_percentage_categorical = missing_number_categorical / len(df)
missing_info_categorical = pd.concat([missing_number_categorical, missing_percentage_categorical], axis=1, keys=['missing number', 'missing percentage'])
missing_info_categorical
```

```python
# replace missing categorical values variables with the mode of each column

df_categorical_mode = df[df_categorical].fillna(df[df_categorical].mode().iloc[0])
df_categorical_mode.head()
```

# Count plot for bivariate analysis

```python
# sns.set_style("darkgrid")
# plt.xticks(rotation=90)
# sns.barplot(x='Neighborhood', y=df['SalePrice'], data=df[df_categorical])

for i in range(0, len(df_categorical), 2):
    plt.figure(figsize=(10, 4))
    plt.subplot(121)
    plt.xticks(rotation=90)
    sns.barplot(y=df['SalePrice'], x=df_categorical[i], data=df_categorical_mode)
    plt.subplot(122)
    plt.xticks(rotation=90)
    sns.barplot(y=df['SalePrice'], x=df_categorical[i + 1], data=df_categorical_mode)
    plt.tight_layout()
    plt.show()
```

# 4C Identify significant variables using p-values and Chi-Square values


```python
df_categorical_mode
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df_categorical_mode.columns:
    le = LabelEncoder()
    df_categorical_mode[col] = le.fit_transform(df_categorical_mode[col])
df_categorical_mode.head()
```

```python
# price = df['SalePrice']
# price
# df_categorical_mode_price = df_categorical_mode.assign(SalePrice=price)
# df_categorical_mode_price.head()
```

```python
from sklearn.feature_selection import chi2
X = df_categorical_mode
y = df['SalePrice']
```

```python
chi_score = chi2(X, y)
chi_score
```

```python
# higher the chi value, higher the importance
chi_values = pd.Series(chi_score[0], index=X.columns)
chi_values.sort_values(ascending=False, inplace=True)
chi_values.plot.bar()
```

```python
# higher the p value, lower the importance
p_values = pd.Series(chi_score[1], index=X.columns)
p_values.sort_values(ascending=False, inplace=True)
p_values.plot.bar()
```

```python
# We will select categorical variables lower than 0.5
selected_significant_var = ['Neighborhood', 'BldgType', 'HeatingQC', 'GarageType', 'LandSlope', 'HouseStyle', 'Exterior2nd', 'LotShape', 'BsmtFinType1', 'Exterior1st', 'Alley']
significant_categorical_var = df_categorical_mode[selected_significant_var]
significant_categorical_var
```
